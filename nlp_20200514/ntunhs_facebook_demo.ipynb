```python
import pandas as pd
from selenium import webdriver
import time
from bs4 import BeautifulSoup

driver = webdriver.Chrome(executable_path = "chromedriver_win32.exe")

driver.implicitly_wait(3)
```

```python
driver.get('https://www.facebook.com/pg/WHO/posts/')
driver.implicitly_wait(10)
time.sleep(3)

driver.execute_script('window.scrollTo(0, document.body.scrollHeight);') # scroll window
time.sleep(3)
driver.execute_script('window.scrollTo(0, document.body.scrollHeight);') # scroll window
time.sleep(3)
driver.execute_script('window.scrollTo(0, document.body.scrollHeight);') # scroll window
time.sleep(3)

#expose all hidden links
showMoreBtn = driver.find_elements_by_css_selector("span.text_exposed_hide > \
                                                    span.text_exposed_link > a.see_more_link")
for c in range(len(showMoreBtn)):
    driver.execute_script("arguments[0].click();", showMoreBtn[c])
    driver.implicitly_wait(1)
else:
    pass

htmltext = driver.page_source
soup = BeautifulSoup(htmltext,"lxml")
```

```python
#create dataframe
columns=['date','org','content','like_string','message_string','share_string']
df=pd.DataFrame(columns=columns)
df
```

```python
post_len=len(soup.select('div._1dwg._1w_m._q7o'))

for i in range(post_len):
    #first block
    first=soup.select('div._1dwg._1w_m._q7o')
    #print("first:",len(first))
    #second block
    second=soup.select('div._57w,div._7a9u')
    #print("second:",len(second))
    
    #org
    first_soup = BeautifulSoup(str(first[i]),"lxml")
    org=first_soup.select('h5._7tae._14f3._14f5._5pbw._5vra a')[0].text.strip()
    date=first_soup.select('span.timestampContent')[0].text.strip()
    content=first_soup.select('div._5pbx.userContent._3576')[0].text.strip()

    second_soup = BeautifulSoup(str(second[i]),"lxml")
    #like
    if second[i].text=='讚留言分享':
        like_string="0"
    else:
        second_soup = BeautifulSoup(str(second[i]),"lxml")
        like_string=second_soup.select('span._81hb')[0].text
    #print(like_string)
    
    #message
    if second[i].text=='讚留言分享':
        message_string="0"
    else:
        second_soup = BeautifulSoup(str(second[i]),"lxml")
        message_string=second_soup.select('a._3hg-._42ft')[0].text.strip()
    #print(message_string)
    
    #share
    if second[i].text=='讚留言分享':
        share_string="0"
    else:
        second_soup = BeautifulSoup(str(second[i]),"lxml")
        share_string=second_soup.select('a._3rwx._42ft')[0].text.strip()
    #print(share_string)
    
    input={"date":date,"org":org,"content":content,"like_string":like_string,\
           "message_string":message_string,"share_string":share_string}
    df=df.append(input,ignore_index=True)
```

```python
df.to_csv('who_facebook.csv',header=True,index=False)
```
