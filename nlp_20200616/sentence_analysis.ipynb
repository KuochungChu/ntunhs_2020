```python
import pandas as pd
```

```python
df=pd.read_csv('update.csv')
```

```python
word_pos=[]
for terms in df[df['label']==1]['pos'].apply(lambda x: x.split(',')):
    for term in terms:
        word_pos.append(term)
```

```python
from collections import Counter
counts = Counter(word_pos)
#print(counts)
```

```python
#找最多字頻
for key, value in counts.most_common():
    print(key, value)
```

```python
#找字詞:死光
for key, value in counts.most_common():
    if key.split('-')[0]=='死光':
        print(key, value)
```

```python
from ckiptagger import data_utils
from ckiptagger import WS, POS, NER
ws=WS("C:/Users/Lenovo/Documents/code/fb_Carrefour/data")
pos=POS("C:/Users/Lenovo/Documents/code/fb_Carrefour/data")
ner=NER("C:/Users/Lenovo/Documents/code/fb_Carrefour/data")
```

```python
def about_ner(text):
    
    if str(text)=='nan':
        return ''
    
    ws_results=ws([text])
    pos_results=pos(ws_results)
    
    res=[]
    for i in zip(ws_results[0], pos_results[0]):
            res.append(i[0]+"-"+i[1])
    return res
```
