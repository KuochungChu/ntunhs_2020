## Correlation Matrix

```python
def plot_corr(df,size=10):
    '''Function plots a graphical correlation matrix for each pair of columns in the dataframe.

    Input:
        df: pandas DataFrame
        size: vertical and horizontal size of the plot'''

    corr = df.corr()
    fig = plt.figure()
    ax = fig.add_subplot(111)
    
    cax = ax.matshow(corr, cmap=plt.cm.Greens)
            
    fig.colorbar(cax)

    plt.xticks(range(len(corr.columns)), corr.columns, rotation=90);
    plt.yticks(range(len(corr.columns)), corr.columns);
    
    for (i, j), z in np.ndenumerate(corr):
        ax.text(j, i, '{:0.1f}'.format(z), ha='center', va='center')

    plt.show()
```

```python
plot_corr(movie_df)
plt.close()
```

```python
movie_df["profit"] =  movie_df["gross"] - movie_df["budget"]
movie_df["ROI"] = ((movie_df["gross"] - movie_df["budget"]) / movie_df["budget"] * 100).fillna(0)
```

```python
most_20_profit_df = movie_df.sort_values(by="profit", ascending=False).head(20)

plt.plot(most_20_profit_df["budget"]/1000000, most_20_profit_df["profit"]/1000000, 'ko')


for txt, xy in zip(most_20_profit_df["movie_title"], zip(most_20_profit_df["budget"]/1000000-10, most_20_profit_df["profit"]/1000000+5)):                                      
    if xy[1] > 300:
        plt.annotate('%s' % txt, xy=xy, textcoords='data') 
    else:
        xy = list(xy)
        xy[1]-=5
        xy = tuple(xy)
        
        if xy[0] > 100:
            xy = list(xy)
            xy[0]+=xy[0]/10
            xy = tuple(xy)
        elif xy[0] > 80:
            xy = list(xy)
            xy[0]+=xy[0]/10
            xy[1]-=12
            xy = tuple(xy)
            if xy[1] < 300 and xy[1] > 272:
                xy = list(xy)
                xy[1]+=13
                xy = tuple(xy)  
        elif xy[0] < 60 and xy[0] > 50  :
            xy = list(xy)
            xy[0]-=10
            xy = tuple(xy)
        elif xy[0] < 70 and xy[0] > 60  :
            xy = list(xy)
            xy[0]+=10
            xy[1]+=10
            xy = tuple(xy)
        else:
            xy = list(xy)
            xy[0]-=18
            xy[1]-=10
            xy = tuple(xy)
        plt.annotate('%s' % txt, xy=xy, textcoords='data')

plt.xlabel('Budget $million')
plt.ylabel('Profit $million')
plt.title('20 Most Profitable Movies')
plt.grid(True)
plt.grid(color='k', linestyle='--', linewidth=1)
plt.plot([0,250],[250,500],ls="-",c="b")

plt.show()
plt.close()
```

```python
from random import randint

def get_text_positions(x_data, y_data, txt_width, txt_height):
    a = zip(x_data, y_data)
    text_positions = y_data.copy()
    for index, (z, (y, x)) in zip(y_data.index,enumerate(a)):
        local_text_positions = [i for i in a if i[0] > (y - txt_height) 
                            and (abs(i[1] - x) < txt_width * 2) and i != (y,x)]
        if local_text_positions:
            sorted_ltp = sorted(local_text_positions)
            if (abs(sorted_ltp[0][0] - y) < txt_height) or (abs(sorted_ltp[0][1] - x) < txt_width): #True == collision
                differ = np.diff(sorted_ltp, axis=0)
                a[z] = (sorted_ltp[-1][0] + txt_height, a[z][1])
                text_positions[index] = sorted_ltp[-1][0] + txt_height
                for k, (j, m) in enumerate(differ):
                    #j is the vertical distance between words
                    if j > txt_height * 2: #if True then room to fit a word in
                        a[z] = (sorted_ltp[k][0] + txt_height, a[z][1])
                        text_positions[index] = sorted_ltp[k][0] + txt_height
                        break
    return text_positions

def text_plotter(x_data, y_data, text_positions, axis,txt_width,txt_height, txt_content):
    for x,y,t,txt_content in zip(x_data, y_data, text_positions, txt_content):
        
        if y != t:
            temp = randint(0,130)
            axis.text(x - txt_width*2, 1.01*temp, '%s'%(txt_content),rotation=0, color='blue')
            axis.arrow(x - txt_width*1.5, temp, +txt_width*1.5, y-temp, \
                       color='red', \
                       alpha=0.3, \
                       zorder=0, \
                       width=0.01, \
                       head_width=0, \
                       length_includes_head=True \
                       )

        else:
            axis.text(x, 1.01*t, '%s'%(txt_content),rotation=0, color='blue')
```

```python
most_20_ROI_df = movie_df.sort_values(by="ROI", ascending=False).head(20)

#random test data:
x_data = most_20_ROI_df["budget"]/1000000
y_data = most_20_ROI_df["profit"]/1000000

#GOOD PLOT:
fig2 = plt.figure()
ax2 = fig2.add_subplot(111)
ax2.plot(x_data, y_data, 'ko')
#set the bbox for the text. Increase txt_width for wider text.
txt_height = 0.3*(plt.ylim()[1] - plt.ylim()[0])
txt_width = 0.3*(plt.xlim()[1] - plt.xlim()[0])

# print (txt_height, txt_width)
#Get the corrected text positions, then write the text.
text_positions = get_text_positions(x_data, y_data, txt_width, txt_height)

text_plotter(x_data, y_data, text_positions, ax2, txt_width, txt_height, most_20_ROI_df["movie_title"])

plt.xlim([-2,4])
plt.xlabel('Budget $million')
plt.ylabel('Profit $million')
plt.title('20 Most ROI Movies')
plt.grid(True)
plt.grid(color='k', linestyle='--', linewidth=1)
plt.plot([0,4],[0,200],ls="-",c="b")

plt.show()
plt.close()
```

```python
use_columns = list(movie_df.columns.values)
use_columns = [n for n in use_columns if n not in \
               ["movie_imdb_link", "genres", "movie_title", "plot_keywords",\
               "actor_1_name", "actor_2_name", "actor_3_name"]]
```

```python
movie_df['color'] = movie_df['color'].astype('category')
movie_df['director_name'] = movie_df['director_name'].astype('category')
movie_df['country'] = movie_df['country'].astype('category')
movie_df['content_rating'] = movie_df['content_rating'].astype('category')
movie_df['language'] = movie_df['language'].astype('category')
```

```python
X_train_df = movie_df[use_columns][:int(0.8*len(movie_df))].fillna(0)
X_test_df = movie_df[use_columns][int(0.8*len(movie_df)):].fillna(0)
```

```python
Y_train_df = movie_df[["imdb_score"]][:int(0.8*len(movie_df))].fillna(0)
Y_test_df = movie_df[["imdb_score"]][int(0.8*len(movie_df)):].fillna(0)
```

## Feature Importances

```python
# Build a forest and compute the feature importances
clf = ensemble.ExtraTreesRegressor(n_estimators=250,
                              random_state=0)

clf.fit(X_train_df, Y_train_df)
importances = clf.feature_importances_
std = np.std([clf_tree.feature_importances_ for clf_tree in clf.estimators_],
             axis=0)
indices = np.argsort(importances)[::-1]

# Print the feature ranking
print("Feature ranking:")

for f in range(X_train_df.shape[1]):
    print("%d. feature %d: %s (%f)" % (f + 1, indices[f], X_train_df.columns[indices[f]], importances[indices[f]]))

# Plot the feature importances of the forest
plt.figure()
plt.title("Feature importances")
plt.bar(range(X_train_df.shape[1]), importances[indices],
       color="r", yerr=std[indices], align="center")
plt.xticks(range(X_train_df.shape[1]), indices)
plt.xlim([-1, X_train_df.shape[1]])
plt.show()
plt.close()
```

## KFold

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
```

```python
# These are "Class objects". For each Class, find the AUC through
# 10 fold cross validation.
X_train_df = movie_df[use_columns][:].fillna(0)
Y_train_df = movie_df[["imdb_score"]][:].fillna(0)

numFolds = 5
kf = KFold(numFolds, shuffle=True)
conv_X = pd.get_dummies(X_train_df)
    
# Models = [LogisticRegression, RandomForestClassifier, SVC]
# params = [{}, {}, {"probability": True}]

Models = [neighbors.KNeighborsRegressor, tree.DecisionTreeRegressor, ensemble.ExtraTreesRegressor]
params = [{}, {}, {"n_estimators":250,"random_state":0}]
```

```python
for Model, param in zip(Models, params):
    total = 0
    for train_indices, test_indices in kf.split(X_train_df):

        # Get the dataset; this is the way to access values in a pandas DataFrame
        train_X = conv_X.loc[train_indices, :]; 
        train_Y = Y_train_df.loc[train_indices]
        test_X = conv_X.loc[test_indices, :]; 
        test_Y = Y_train_df.loc[test_indices]

        # Train the model, and evaluate it
        clf = Model(**param)
        clf.fit(train_X, train_Y)
        predictions = clf.predict(array(test_X))[:]
        score = metrics.mean_squared_error(test_Y, predictions)
        total += score
    MSE = total / numFolds
    print ("MSE of {0}: {1}".format(Model.__name__, MSE))
    
#         fpr, tpr, _ = metrics.roc_curve(test_Y, predictions)
#         total += auc(fpr, tpr)
#     accuracy = total / numFolds
#     print "AUC of {0}: {1}".format(Model.__name__, accuracy)
```
